### Customer Segmentation using K-Means

```python
# Performing segmentation based on two features -> Annual Income and Spending Score

kmc_df = df.iloc[:,3:]
```

```python
#Scaling the data

scaler = StandardScaler()

df_scaled = scaler.fit_transform(kmc_df)
```

```python
#Finding the optimal number of K using the Elbow Method

wcss = []

for i in range(2,20):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)
    

#Plotting the Scores

plt.figure(figsize = (9,6))

plt.plot(wcss, "bx-")
plt.title("Finding the optimal number of K")
plt.xlabel("Clusters")
plt.ylabel("Scores WCSS")
plt.xticks(np.arange(0,21,1));
```

**Based on the Elbow Method, we should cluster our data into 3 different groups**

```python
#Instantiate and fit KMeans

kmeans = KMeans(n_clusters = 3)

cluster_predictions = kmeans.fit_predict(df_scaled)

labels = kmeans.labels_
```

```python
#Merge the clusters labels to our dataframe

df_cluster = pd.concat([kmc_df, pd.DataFrame({"cluster": labels})], axis = 1)

df_cluster.head(10)
```

```python
#Plotting the histogram for all the clusters

for i in kmc_df.columns:
    plt.figure(figsize = (35,5))
    for j in range (3):
        plt.subplot(1,4, j+1)
        cluster = df_cluster[df_cluster["cluster"] == j]
        cluster[i].hist(bins = 20)
        plt.title("{} \nCluster {}".format(i,j))
```

```python
#Visualizing the clusters

plt.figure(figsize = (9,9))

sns.scatterplot(data = df_cluster,
                x = "Annual Income (k$)",
                y = "Spending Score (1-100)",
                hue = "cluster",
                palette = ["red", "green", "blue"]);
```

**Using K-Means Clustering, we segmented the mall customers in 3 different groups with distinct characteristics:**

**Cluster 0 -> Medium/High Income and Medium/High Spending Score**

**Cluster 1 -> Medium/High Income and Low Spending Score**

**Cluster 2 -> Low/Medium Income and Medium Spending Score**

-----------------------------

**We can conclude that Cluster 0 is pointing to the most attractive customers from a business perspective, while the Cluster 2 has potential for conversion (since they have good purchasing power).**

**Regarding Cluster 3, it is the least attractive customer group, meaning that the mall shouldn't invest many marketing efforts in this segment.**
